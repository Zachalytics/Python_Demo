```python
#ok next is the busy days bit, first stop, finding the days that are busy
#remember the weeks starts at sunday as 0
df_opm['order_day_of_week'].value_counts(dropna=False)
```

```python
#ok cool, sunday is the big day, followed by monday. The slowest is Thursday.
#part of me wants to rename the days but it's probably best I don't for now.
result=[] #this sets up the results so I can port them in later nice and easy
for value in df_opm['order_day_of_week']:
    if value == 0:
        result.append('Busiest Day')
    elif value == 4:
        result.append('Least Busy')
    else:
        result.append('Regular Busy')
```

```python
#now for that easy fill
df_opm['busiest day']=result
#then we check the frequencies
df_opm['busiest day'].value_counts()
```

```python
#so busiest and least both match the results for 0 and 4, cool beans
```

```python
#going to add in the values as ==. I could probably do ranges in this case but I want to keep the days seperated
result2=[] 
for value in df_opm['order_day_of_week']:
    if value == 0:
        result2.append('Busiest Day')
    elif value == 1:
        result2.append('Second Busiest Day')
    elif value == 3:
        result2.append('Second Slowest Day')
    elif value == 4:
        result2.append('Slowest Day')
    else:
        result2.append('Regular Busy')
```

```python
#add it in
df_opm['busiest_days']=result2
#frequency check
df_opm['busiest_days'].value_counts(dropna=False)
```

##### looks like it worked out. Busiest and second busiest match up with 0 and one, slowest and second slowest match up with 4 and 3 respectivly. Cool beans
