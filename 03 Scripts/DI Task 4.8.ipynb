### #6
##### in which we locate some big spenders

```python
#ok grouping by id and fining the mean of the money spent and putting that in it's own column
df['mean_spent']=df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#as per usual, checking to see if it worked
df[['user_id','mean_spent']].head(20)
```

```python
#now we set the ranges. >10=high spender, <10 = low spender
df.loc[df['mean_spent']<5,'spender_level']='Low Spender'
```

```python
df.loc[df['mean_spent']>=5,'spender_level']='High Spender'
```

```python
#check check
df[['user_id','mean_spent','spender_level']].head(20)
```

### #7
##### in which we realize just how much you can tell about someone by their spending habits...

```python
#ok, now we set up a column for the median of the apps usage
df['median_usage']=df.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
#double check that we have numbers
df['median_usage'].head()
```

```python
#ok those are all the same and that makes me nervous
df['median_usage'].head(30)
```

```python
#ok cool. Next step we asign the flags
df.loc[df['median_usage']>20,'customer_frequency']='Non-Frequent Customer'
df.loc[(df['median_usage']<=20)&(df['median_usage']>10),'customer_frequency']='Regular Customer'
df.loc[df['median_usage']<=10,'customer_frequency']='Frequent Customer'
```

```python
#cheaky checky
df[['user_id','median_usage','customer_frequency']].head(30)
```
